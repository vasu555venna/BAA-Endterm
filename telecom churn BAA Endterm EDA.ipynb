```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
vvv = tel['gender'].value_counts()
sns.set(style="darkgrid")
sns.barplot(vvv.index, vvv.values, alpha=0.9)
plt.title('Frequency Distribution of Churn rate based on gender')
plt.ylabel('Churn', fontsize=12)
plt.xlabel('gender', fontsize=12)
plt.show()
```
